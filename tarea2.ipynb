```python
#EJERCICIO 3

num1 = int(input("Ingresa el primer número: "))
num2 = int(input("Ingresa el segundo número: "))

match num1:
    case _ if num1 < num2:
        print("El número 2 es mayor")
    case _ if num1 > num2:
        print("El número 1 es mayor")
    case _:
        print("Los dos son iguales")
```

```python
# EJERCICIO 4
num1 = int(input("Ingresa el primer número: "))
num2 = int(input("Ingresa el segundo número: "))

inicio = min(num1, num2)
fin = max(num1, num2)

suma_total = 0

for i in range(inicio, fin + 1):
    suma_total += i

print("La suma de los enteros desde", inicio, "hasta", fin, "es:", suma_total)
```

```python
bajo = int(input("Ingresa el límite inferior: "))
alto = int(input("Ingresa el límite superior: "))
paso = int(input("Ingresa el valor de paso (step): "))
print(f"\nIterando de {bajo} a {alto} con paso {paso}:")
for i in range(bajo, alto + paso, paso):
    if i <= alto:
        print(i)
```
