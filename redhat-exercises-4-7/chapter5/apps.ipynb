```python
import sys

argumentos = sys.argv[1:]
argumentos_ordenados = sorted(argumentos)

for arg in argumentos_ordenados:
    print(arg)
```

```python
import sys

argumentos = sys.argv[1:]
numeros = []

for arg in argumentos:
    try:
        numeros.append(float(arg))
    except ValueError:
        pass

if len(numeros) > 0:
    suma = sum(numeros)
    promedio = suma / len(numeros)
    
    print(suma)
    print(promedio)
```
