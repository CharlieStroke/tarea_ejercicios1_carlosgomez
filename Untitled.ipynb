```python
#EJERCICIO 2
def obtener_iniciales(nom, ape):
    inicial1 = nom[0].upper()
    inicial2 = ape[0].upper()
    return inicial1 + inicial2

nombre = input()
apellido = input()

print (nombre, apellido)
print (obtener_iniciales(nombre, apellido))
```

```python
#EJERCICIO 3
def haypunto(cadena):
    return cadena.endswith(".")

def solo_alfabeto(cadena):
    return cadena.isalpha()

def hayx(cadena):
    return "x" in cadena
    
def e_upper(cadena):
    nueva_cadena = cadena.replace("e", "E")
    return nueva_cadena;
    

texto = input()

if haypunto(texto):
    print("El texto termina con punto")
else:
    print("El texto no termina con punto")
    
if solo_alfabeto(texto):
    print("El texto contiene solo letras del alfabeto")
else:
    print("El texto contiene espacios, numeros, simbolos y demas caracteres fuera del alfabeto")

if hayx(texto):
    print ("Hay X en la cadena")
else:
    print ("No hay X en la candena")

if "e" in texto:
    print (e_upper(texto))
else:
    print ("No hay e's en la cadena")
```

```python
#EJERCICIO 4
def funcion(cadena):
    if not cadena:
        return "cadena vacia"
    inicial = cadena[0]
    final = cadena[-1]
    print ('La inicial es ', inicial.upper(), 'y se repite ', cadena.lower().count(inicial.lower()), 'vez/veces')
    print ('Finaliza con ', final.upper(), 'y se repite ', cadena.lower().count(final.lower()), 'vez/veces')    

texto = input()
funcion(texto)
```
